# TOP LOSSER AND GAINER WEB SCRAPPING FROM NSE

```python

import bs4
import requests
import urllib
import json
from bs4 import BeautifulSoup
import pprint
```

```python
stock_url="https://www.nseindia.com/#gainers_loosers"
headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.1.1 Safari/605.1.15","Accept-Language": "en-gb","Accept-Encoding":"br, gzip, deflate","Accept":"test/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8","Referer":"http://www.google.com/"}

response = requests.get(stock_url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
#print(soup)
```

```python
ScriptList=soup.find_all('script')

#ScriptList
```

```python
ScriptList[1].getText().strip()
```

```python
data_array = ScriptList[1].getText().strip()
type(data_array)
```

```python
len(data_array)
```
